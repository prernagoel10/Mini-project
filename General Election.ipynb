With over 600 Million voters voting for 8500+ candidates across 543 constituencies, the general elections in the world's largest democracy are a potential goldmine of data. # Graph-STATE VS CANDIDATE

```python
plt.figure(figsize=(20,10))
sms.set(style="dark")
a=sms.countplot(x="STATE",data=File)
plt.grid()
plt.title('Candidate by State')
a=a.set_xticklabels(a.get_xticklabels(),rotation=60)
```

# GRAPH BETWEEN TOTAL VOTES VS VOTES
This graph will show the No of votes vs votes.


```python
a= File['TOTAL\nVOTES'].sum()     
b=a+File['GENERAL\nVOTES'].sum()
mains=["VOTES","TOTAL VOTERS"]
quantity=[a,b]
plt.pie(quantity,labels=mains,radius=2,autopct="%0.1f%%",colors=["yellow","blue"])
plt.pie([b],colors=["white"],radius=1)
plt.show()
```

# OBESERVATIONS
This graph show the Total number of votes vs Total votes.
